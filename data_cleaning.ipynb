```python
import os
import pandas as pd

bike_2017_csv = "Divvy_Trips_2017.csv"
bike_2017_df = pd.read_csv(bike_2017_csv)

bike_2018_csv = "Divvy_Trips_2018.csv"
bike_2018_df = pd.read_csv(bike_2018_csv)

CTA_csv = "CTA_Daily_Boarding_Totals.csv"
CTA_df = pd.read_csv(CTA_csv)
```

```python
bike_2017_df.shape
```

```python
#Create new DF with columns needed
reduced_bike_2017_df = bike_2017_df.loc[:,["trip_id", "starttime"]]

reduced_bike_2018_df = bike_2018_df.loc[:,["trip_id", "starttime"]]
```

```python
#Check for null values
reduced_bike_2017_df.isnull().any()
reduced_bike_2018_df.isnull().any()
```

```python
#Rename columns
clean_2017_df = reduced_bike_2017_df.rename(index=str, columns={"trip_id": "Trip ID", "starttime": "Date"})

clean_2018_df = reduced_bike_2018_df.rename(index=str, columns={"trip_id": "Trip ID", "starttime": "Date"})
```

```python
clean_2017_df['Date'] = pd.to_datetime(clean_2017_df.Date)
clean_2017_df.sort_values(by="Date")

clean_2018_df['Date'] = pd.to_datetime(clean_2018_df.Date)
clean_2018_df.sort_values(by="Date")
```

```python
new_clean_2017 = clean_2017_df.groupby(["Date"])["Trip ID"].count().reset_index(name="Total Number of Daily Rides")


new_clean_2018 = clean_2018_df.groupby(["Date"])["Trip ID"].count().reset_index(name="Total Number of Daily Rides")

new_clean_2018.head()
```

```python
combined_df = new_clean_2017.append(new_clean_2018, ignore_index=True)


combined_df = combined_df.rename(index=str, columns={"Date": "Bike Rental Date"})
combined_df
```

```python
combined_df.to_csv('2017_2018_Divvy_Trips.csv')
```

```python
CTA_df = CTA_df.rename(index=str, columns={"service_date": "Date", "Bus": "Bus Boardings", "rail_boardings": "Rail Boardings",
                                          "Total Number of CTA Rides": "Total Number of Boardings"})
CTA_df.head()
```

```python
CTA_df['Date'] = pd.to_datetime(CTA_df.Date)
CTA_df.sort_values(by="Date")
```

```python
clean_cta_df = CTA_df.drop(['day_type'], axis=1)

clean_cta_df = clean_cta_df[(clean_cta_df.Date > "2016-12-31")]
clean_cta_df
```

```python
clean_cta_df.to_csv('CTA_Total.csv')
```
