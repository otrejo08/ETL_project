```python
import os
import pandas as pd

bike_2017_csv = "Divvy_Trips_2017.csv"
bike_2017_df = pd.read_csv(bike_2017_csv)

bike_2018_csv = "Divvy_Trips_2018.csv"
bike_2018_df = pd.read_csv(bike_2018_csv)

bike_2017_df.head()
```

```python
bike_2017_df.shape
```

```python
#Create new DF with columns needed
reduced_bike_2017_df = bike_2017_df.loc[:,["trip_id", "starttime"]]
reduced_bike_2017_df.head()
```

```python
#Check for null values
reduced_bike_2017_df.isnull().any()
```

```python
#Rename columns
clean_2017_df = reduced_bike_2017_df.rename(index=str, columns={"trip_id": "Trip ID", "starttime": "Date"})
clean_2017_df.head()
```

```python
clean_2017_df['Date'] = pd.to_datetime(clean_2017_df.Date)
clean_2017_df.sort_values(by="Date")
```

```python
new_clean_2017 = clean_2017_df.groupby(["Date"])["Trip ID"].count().reset_index(name="Total Number of Daily Rides")

new_clean_2017.head()
```
